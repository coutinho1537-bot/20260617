# 라이브러리 설치

다음 명령어로 파이썬 가상환경을 설치한다.

python -m venv .venv

python -m pip install ipykernel
python -m pip install ipywidgets

python -m pip install upgrade pip

.\.venv\Scripts\Activate.ps1
